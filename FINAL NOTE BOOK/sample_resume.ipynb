```python
data_1 = pd.DataFrame(data=resume_content, columns=['Resumes'])

# Check the length of the DataFrame
data_length = len(data_1)

# Check the length of the labels list
labels_length = len(labels)

# Adjust the lengths if necessary
if labels_length > data_length:
    labels = labels[:data_length]
elif labels_length < data_length:
    labels = labels + ['Unknown'] * (data_length - labels_length)

# Assign the labels to the 'category' column
data_1['category'] = labels

# Remove rows where the category is 'Unknown'
data_1 = data_1[data_1['category'] != 'Unknown']
data_1
```
